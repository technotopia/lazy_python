Программа ниже считает окружность Земли, первую космическую скорость (при которой спутник не падает на планету), время одного оборота вокруг Земли. Вычисления для случая низкой околоземной орбиты.

Про первую космическую скорость. Она получается, когда спутник не падает, а значит ускорение свободного падения уравновешено ускорением центробежной силы. 

$$ a = \frac{v^2}{r} , v = \sqrt{ar}, a = g$$

```python
import math

PI = math.pi

radius = 6400000 # радиус Земли в метрах (СИ)
planet = "Земля"

equator = 2 * PI * radius
print("длинна окружности экватора планеты %s - %.2f м или %.2f км" % (planet, equator, equator / 1000))
#На самом деле экватор длиннее, так как Земля не совсем сфера, она немного сплющена из-за вращения.

g = 10.0 

v1 = math.sqrt(g * radius)

print("первая космическая скорость для планеты %s - %.2f м/c или %.2f км/c" % (planet, v1, v1 / 1000))

orbit_time = radius / v1

print("спутник сделает оборот вокруг планеты %s за %.0f секунд" % (planet, orbit_time))
```

Вообще-то первая космическая для Земли не ровно 8 км/c, а чуточку меньше - 7,9 км/c. У нас примерные данные, если взять более точные входные цифры, то результат будет лучше. Вот табличка внизу, попробуйте подставить числа из нее.
