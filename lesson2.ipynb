Решатель квадратных уравнений, условные операторы.

Лень - двигатель прогресса. Программы нужны, чтобы делать жизнь проще и автоматизировать скучные однообразные задачи. Вот взять хотя бы квадратные уравнения - решать их приходится постоянно. Есть общая формула корней квадратного уравнения через дискриминант.

$$a x^2 + b x + c = 0 \quad  x^2 + \frac{b}{a} x + \frac{c}{a} = 0$$

$$ (x + \frac{b}{2a})^2 = \frac{b^2}{4a^2} - \frac{c}{a}$$

$$ x = \frac{-b \pm \sqrt{b^2 - 4ac} }{2a}$$

Ничего сложного вроде бы. Но надо сделать кучу арифметических действий, взять корень, ошибиться легко.



Как говорил агент Смит: "*Не надо посылать человека делать работу машины".* Давайте напишем свою решалку квадратных уравнений.



```python
import math

def solve(a,b,c):
    d = b * b - 4 * a * c
    root1 = (-b + math.sqrt(d)) / (2 * a)
    root2 = (-b - math.sqrt(d)) / (2 * a)
    return root1, root2


solve(1,-4,3)
```

Казалось бы, все очень просто.

Пишем простую функцию, на вход получаем параметры a,b,c из $ax^2 + bx + c = 0$

Посчитали дискриминант, взяли из него квадратный корень, подставили, посчитали оба корня уравнения. Всего делов то! Или нет? 

К сожалению, нет. Ведь квадратное уравнение предоставляет множество способов выстрелить себе в ногу. Самый очевидный способ это сделать:

```python
solve(1,1,1)
```

А как еще можно? Сколько способов можете назвать?

1. Отрицательный дискриминант. Корень из отрицательного числа мы пока брать не умеем. Ничего не будем возвращать, комплексные числа не знаем.
2. $a = 0$ уравненние превратилось в линейное. На ноль делить нельзя. Надо вернуть один корень линейного уравнения. 
3. $a = 0$ и $b = 0$ тут возможны два варианта: корней нет совсем, если с не равно 0 или, если с = 0, то любой x годится, потому что 0 = 0. Впрочем, будем считать, что корней нет.

Так как же в Питоне сделать выбор?

Для этого есть операторы **if** (еще else и elif). 

Маленький пример, функция, которая проверяет знак числа. Выполните код внизу:

```python
def is_positive(num):
    if num > 0:
        print ("число %d - положительное" % num)
    elif num < 0:
        print ("число %d - отрицательное" % num)
    else:
        print ("число %d - ноль" % num)

is_positive(5)
is_positive(-3)
is_positive(0)
```

Давайте разберемся, что это такое было.

**if** проверяет условие **num больше 0** 

если условие выполнено, печатает "положительное"

**elif** - это сокращение от else  if. Если предыдущее условие не выполнено, то может быть выполнено условие **num меньше 0**

В самом конце идет **else** - иначе. Если условия выше не выполнены, выполнится то, что идет после else.

Обратите внимание на двоеточия и отступы. После условия идет не просто команда в блок кода. Чтобы Питон понимал, где начинается и заканчивается блок нужны отступы.

А что вообще такое **num > 0**?

Давайте попросим Питон это распечатать:

```python
print(1 > 0)
print(1 < 0)
```

 Это не строки, а **булевый тип данных** - у него бывает только два значения: **True - истина. False - ложь**.
 
 Выше мы выяснили как узнать, что число отрицательное - пригодится для дискриминанта и квадратного корня. А как проверить, что число не равно нулю - чтобы случайно на него не поделить?
 
 Операторы сравнения:
 
 | оператор | что проверяет | примеры |
 | --- | ---| --- |
 | a < b | a меньше b | (3 < 4) - True, (7 < 7) - False |
 | a > b | a больше b | (4 > 3) - True, (7 > 7) - False |
 | a == b | a равно b | (4 == 4) - True, (7 == 8) - False |
 | a != b | a не равно b | (4 != 3) - True, (7 != 7) - False |
 | a >= b | a больше или равно b | (4 >= 3) - True, (7 >= 7) - True |
 | a <= b | a меньше или равно b | (3 <= 4) - True, (7 <= 7) - True |

```python
# notice the difference between = and ==
# operator = assigns value (on the right) to the variable (on the left)
# operator == checks if left and right are equal and returns boolean
print("1 == 0 is %r" % (1 == 0))
print("0 == 0 is %r" % (0 == 0))
# operator != checks that left and right are not equal 
print("1 != 0 is %r" % (1 != 0))
print("0 != 0 is %r" % (0 != 0))
# >= means greater or equal, <= less or equal
print ("7 >= 7 is %r" % (7 >= 7))
print ("8 <= 8 is %r" % (8 <= 8))
```

```python
import math

def div(a,b):
    if (b != 0):
        return a / b
    else:
        return "can't divide by zero"
    
print (div(9,3))
print (div(7,2))
print (div(6,0))

def sqrt(a):
    if (a >= 0):
        return math.sqrt(a)
    else:
        return "can't take square root of negative number"
    
print (sqrt(16))
print (sqrt(25))
print (sqrt(-1))
```

Вот заготовка функции для вычисления корней квадратного уравнения. Подставьте в нее нужные проверки.

Сейчас в условиях подставлены True или False, их надо заменить на операторы сравнения.



```python
def solve(a,b,c):
    d = b * b - 4 * a * c
    if a == 0: # case of a = 0, so bx + c = 0
        if b != 0: # b is not 0, can divide
            return -c / b        
    elif d >= 0: # a is not zero, discriminant is >= 0
        root1 = (-b + math.sqrt(d)) / (2 * a)
        root2 = (-b - math.sqrt(d)) / (2 * a)
        return root1, root2

assert solve(1,-5,6) == (3., 2.), "неправильное решение с обычными корнями"
assert solve(1,2,1) == (-1., -1.), "неправильное решение с обычными корнями"
assert solve(1,1,1) == None, "неправильное решение с отрицательным дискриминантом"
assert solve(0,1,8) == (-8.), "неправильное решение с a = 0"
assert solve(0,0,8) == None, "неправильное решение с a = 0, b = 0"    
    
        

        
    
```
